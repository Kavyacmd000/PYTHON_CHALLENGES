Tuples (Creation, Modifi cation and Access) :
Create 1st tuple with values -> (10, 20, 30), 2nd tuple with values -> (40, 50, 60):

```python
tup1=(10,20,30)
tup2=(40,50,60)
print(tup1)
print(tup2,"\n")

print(" a. Concatenate the two tuples and store it in “t_combine” ")
t_combine=tup1+tup2
print(t_combine,"\n")

print(" b. Repeat the elements of “t_combine” 3 times ")
print(t_combine*3,"\n")

print(" c. Access the 3rd element from “t_combine” ")
print(t_combine[2],"\n")

print(" d. Access the first three elements from “t_combine” ")
print(t_combine[0:3],"\n")

print(" e. Access the last three elements from “t_combine” ")
print(t_combine[-3:],"\n")
```
